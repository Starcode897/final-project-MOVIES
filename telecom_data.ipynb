```python
import pandas as pd
#import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report
```

```python
path = "resources/telecom_users.csv"
df = pd.read_csv(path)
df
```

```python
#Turn unrecognized spaces into NaNs
df['TotalCharges'].replace(' ', np.nan, inplace=True)
df['MonthlyCharges'].replace(' ', np.nan, inplace=True)
```

```python
df.info()
```

```python
#Find Strings
df[pd.to_numeric(df.TotalCharges, errors='coerce').isnull()]
#It's a space. Erase the spaces.

#Find Numbers
#df[pd.to_numeric(df.TotalCharges, errors='coerce').notnull()]
```

```python
#Drop fake NaNs because dropna doesn't recognize these
df.dropna(subset=['TotalCharges'], inplace=True)
df
```

```python
#NaNs are gone
df[pd.to_numeric(df.TotalCharges, errors='coerce').isnull()]
```

```python
df["TotalCharges"] = df["TotalCharges"].astype(float)
df.to_csv("resources/telecom_cleaned.csv", index=False)
```

```python
#inverse_transform(y) <-- How to turn labels back
label = LabelEncoder()

#df["MonthlyCharges"] = label.fit_transform(df["MonthlyCharges"])
#df["TotalCharges"] = df["TotalCharges"].round(1)
#df["MonthlyCharges"] = df["MonthlyCharges"].round(1)

infos = ["gender", "SeniorCitizen", "Partner", "Dependents", "tenure", "PhoneService", 
        "MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection",
        "TechSupport", "StreamingTV", "StreamingMovies", "Contract", "PaperlessBilling", "PaymentMethod", "Churn",
        ]


for info in infos:
    df[f"{info}"] = label.fit_transform(df[f"{info}"])
    #df["gender"].value_counts()
    
```

```python
data = df[["gender", "SeniorCitizen", "Partner", "Dependents", "tenure", "PhoneService", 
           "MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection",
          "TechSupport", "StreamingTV", "StreamingMovies", "Contract", "PaperlessBilling", "PaymentMethod",
          "TotalCharges", "Churn"]]
data
```

```python
data.info()
```

```python
X = data
y = df["MonthlyCharges"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
y_train
```

```python
sc =  StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

X_train
```

```python
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train, y_train)
pred_rfc = rfc.predict(X_test)
print(pred_rfc)
print(y_test)
```

```python
print(f"Training Data Score: {rfc.score(X_train, y_train)}")
print(f"Testing Data Score: {rfc.score(X_test, y_test)}")
```

```python
print(classification_report(y_test, pred_rfc))
print(confusion_matrix(y_test, pred_rfc))
```
