```python
path = "resources/telecom_users.csv"
df = pd.read_csv(path)
df
```

```python
#Turn unrecognized spaces into NaNs
df['TotalCharges'].replace(' ', np.nan, inplace=True)
df['MonthlyCharges'].replace(' ', np.nan, inplace=True)
```

```python
df.info()
```

```python
#Find Strings
df[pd.to_numeric(df.TotalCharges, errors='coerce').isnull()]
#It's a space. Erase the spaces.

#Find Numbers
df[pd.to_numeric(df.TotalCharges, errors='coerce').notnull()]
```

```python
#Drop NaNs
df.dropna(subset=['TotalCharges'], inplace=True)
df
```

```python
#NaNs are gone
df[pd.to_numeric(df.TotalCharges, errors='coerce').isnull()]
```

```python
df["TotalCharges"] = df["TotalCharges"].astype(float)
df.to_csv("resources/telecom_cleaned.csv", index=False)
```
